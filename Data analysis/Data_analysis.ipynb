Rating

```python
rating= fifa[['name', 'overall_rating']]
rating.head()
```

```python
plt.figure(figsize=(10,7))
plt.bar(list(rating['name'])[0:10],list(rating['overall_rating'])[0:10],color="b")
plt.show()
```

Rating in terms of Potential

```python
p_rating= fifa[['name', 'potential']]
p_rating.head()
```

```python
p_rating= p_rating.sort_values(by=['potential'], ascending=False)
p_rating.head(10)
```

Salary

```python
salary= fifa[['name', 'value_euro']]
```

```python
salary.head()
```

```python
salary= salary.sort_values(by=['value_euro'], ascending=False)
```

```python
salary.head(10)
```

```python
plt.figure(figsize=(15,7))
plt.bar(list(salary['name'])[0:10],list(salary['value_euro'])[0:10],color="y")
plt.show()
```
